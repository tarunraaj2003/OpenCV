```python
import cv2,time
```

```python
vid=cv2.VideoCapture(0)
f=1

while True:
    f=f+1
    check,frame=vid.read()
    print(frame)
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("Video Capture",gray)
    x = cv2.waitKey(1)
    if x == ord('p'):
        break

print(f)
vid.release()
cv2.destroyAllWindows()
```
